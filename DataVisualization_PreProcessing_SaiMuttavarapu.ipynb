```python
# Id, Country, Album_Type, Gender, Lifespan_begin,Lifespan_end, Lifespan_ended?

new_df = pd.DataFrame()
```

```python
new_df["ID"] = df["_id"]
new_df["Country"] = df["location.country"]
new_df["Album_Type"] = df["type"]
new_df["Gender"] = df["gender"]
new_df["Lifespan_Begin"] = df["lifeSpan.begin"]
new_df["Lifespan_End"] = df["lifeSpan.end"]
new_df["Lifespan_End?"] = df["lifeSpan.ended"]
```

- **Let us assume the begin year as 1970 for the albums with unknown begin year**
- **Let us assume the end year as 2050 for the albums with no end year**
- **Let us assume the end year as 2021 for the albums with unknown end year**

```python
new_df["Lifespan_Begin"] = np.where(new_df["Lifespan_Begin"].isnull(), 1970, new_df["Lifespan_Begin"])
```

```python
new_df["Lifespan_End"] = np.where(new_df["Lifespan_End?"]==False & new_df["Lifespan_End"].isnull(), 2050, new_df["Lifespan_End"])
new_df["Lifespan_End"] = np.where(new_df["Lifespan_End?"]==True & new_df["Lifespan_End"].isnull(), 2021, new_df["Lifespan_End"])
```

```python
new_df["Gender"].value_counts()
```

```python
new_df["Gender"].isnull().sum()
```

```python
new_df["Gender"] = np.where(new_df["Gender"].isnull(), "Unknown", new_df["Gender"])
```

```python
new_df["Album_Type"] = np.where(new_df["Album_Type"].isnull(), "Unknown", new_df["Album_Type"])
```

```python
new_df["Country"] = np.where(new_df["Country"].isnull(), "Unknown", new_df["Country"])
```

```python
new_df["Lifespan_Begin"] = [int(parser.parse(str(x).replace("????","1970")).year) for x in new_df["Lifespan_Begin"]]
```

```python
new_df["Lifespan_End"] = [int(parser.parse(str(x)).year) for x in new_df["Lifespan_End"]]
```
